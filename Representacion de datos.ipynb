```python
import numpy as np
import matplotlib.pyplot as plt
import re

data = [
    './data_local/salida',
    './data_local/salida_2',
    './data_local/salida_4'
]
labels = ['serial', 'OMP_NUM_THREADS=2','OMP_NUM_THREADS=4']
colors = ['red', 'green', 'blue']
fig, ax = plt.subplots()
for i, x in enumerate(data):
    fname = x
    with open(fname) as f:
        content = f.readlines()
        content = [x.rstrip() for x in content] 
        a = np.array([x for i, x in enumerate(content) if not (i % 4 == 0)])
        a = a.astype(np.double)
        a = a.reshape((30,3))
        a = [ x.sum()/3 for x in a] 
        ax.scatter([x for x in range(100,3100,100)], a, c=colors[i],label=labels[i],  alpha=0.5)

    
```

```python
ax.set_title('Host local')
ax.set_ylabel('Tiempo en milisegundos')
ax.set_xlabel('n varibles')
ax.legend()
ax.grid(True)
plt.show()
```
